**Traduciendo Variables**

```python
datos.columns
```

```python
dic_columnas = {
    'LIMIT_BAL': 'Limite', 
    'CHECKING_ACCOUNT': 'Cuenta_corriente', 
    'EDUCATION': 'Escolaridad', 
    'MARRIAGE': 'Estado_civil', 
    'AGE': 'Edad',
    'BILL_AMT': 'Valor_factura', 
    'PAY_AMT': 'Valor_pago',
    'DEFAULT': 'Moroso'
}
```

```python
tarjetas = datos.rename(columns = dic_columnas)
```

```python
tarjetas.head()
```

```python
tarjetas.Cuenta_corriente.unique()
```

```python
dic_cuenta = {
    'Yes': 'Si',
    'No': 'No'
}
```

```python
tarjetas.Cuenta_corriente.map(dic_cuenta)
```

```python
tarjetas.Cuenta_corriente = tarjetas.Cuenta_corriente.map(dic_cuenta)
tarjetas.head()
```

```python
tarjetas.Escolaridad.unique()
```

```python
dic_escolaridad = {
    '2.University' : '2.Universidad', 
    '3.Graduate School' : '3.Post-graduacion', 
    '1.High School' : '1.Colegio'
}
```

```python
tarjetas.Escolaridad = tarjetas.Escolaridad.map(dic_escolaridad)
tarjetas.head()
```

```python
tarjetas.Estado_civil.unique()
```

```python
dic_civil = {
    'Married' : 'Casado/a', 
    'Single' : 'Soltero/a'
}
```

```python
tarjetas.Estado_civil = tarjetas.Estado_civil.map(dic_civil)
tarjetas.head()
```

**Test de hipotesis**

```python
sns.lmplot(x = 'IU', y = 'Valor_factura', data = tarjetas, col = 'Moroso');
```

```python
from scipy.stats import ranksums
```

H<sup>null<sup>
    
La distribucion de los grupos morosos y no moroso es la misma

H<sup>alt<sup>
    
La distribucion de los grupos morosos y no moroso no es la misma

```python
moroso = tarjetas.query("Moroso == 1").Valor_factura
```

```python
moroso
```

```python
no_moroso = tarjetas.query("Moroso == 0").Valor_factura
```

```python
no_moroso
```

```python
resultado = ranksums(moroso, no_moroso)
print("El p_value de nuestro test es de {}".format(resultado.pvalue))
```
